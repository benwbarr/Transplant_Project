```python
import pandas as pd 
import numpy as np 
```

```python
cPRA = pd.read_csv('CPRA.csv')
cPRA.head()
```

```python
dropped_columns= cPRA.drop(columns=['Unnamed: 2', 'Heart', 'Lung', 'Heart / Lung',
                   'Intestine', 'Abdominal Wall', 'Head & Neck: Craniofacial', 
                   'GU: Uterus', 'Upper Limb: Bilateral', 'Upper Limb: Unilateral', 'Face, Scalp'])
dropped_columns.head()
```

```python
df = dropped_columns.fillna('-')

renamed_df= df.rename(columns={"Unnamed: 0": "Med. Center", "Unnamed: 1": "cPRA Values"})
renamed_df.head()
```

```python
new_index= renamed_df.set_index('Med. Center')

new_index['cPRA Values'] = new_index['cPRA Values'].str.replace('All CPRA', '')

new_index.head(30)
```
