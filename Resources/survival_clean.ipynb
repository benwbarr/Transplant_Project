```python
import pandas as pd
```

-------------------------------------------------------------------------------------------------------------------------------

Age

-------------------------------------------------------------------------------------------------------------------------------

```python
csv_path_serv_age = "Serv_by_Age.csv"

serv_age_df = pd.read_csv(csv_path_serv_age, encoding="utf-8")

serv_age_df.head(50)
```

```python
#Removing strings from Organ Column

serv_age_df = serv_age_df[serv_age_df.Organ != "Heart / Lung"]

serv_age_df = serv_age_df[serv_age_df.Organ != "Intestine"]

serv_age_df = serv_age_df[serv_age_df.Organ != "Lung"]
```

```python
#Fix names

serv_age_df = serv_age_df.rename(columns={"Number Functioning /": "Number_Functioning_or_Alive",
                                         "Recipient": "Recipient_Age",
                                         "Years": "Years_Post_Transplant",
                                         "Survival": "Survival_Rate"})
```

```python
#Select columns and remove first row

serv_age_df = serv_age_df[["Region", "Organ", "Recipient_Age", "Years_Post_Transplant", "Number_Functioning_or_Alive", "Survival_Rate"]]
```

```python
#Remove null values

serv_age_df = serv_age_df[serv_age_df.Region != "NaN"]

serv_age_df = serv_age_df[serv_age_df.Organ != "NaN"]

serv_age_df = serv_age_df[serv_age_df.Recipient_Age != "Age"]

serv_age_df = serv_age_df[serv_age_df.Years_Post_Transplant != "Post Transplant"]

serv_age_df = serv_age_df[serv_age_df.Number_Functioning_or_Alive != "Alive"]

serv_age_df = serv_age_df[serv_age_df.Survival_Rate != "Rate"]

serv_age_df = serv_age_df[serv_age_df.Survival_Rate != "*"]

serv_age_df.head(50)
```

```python
serv_age_df.to_csv("Cleaned/serv_age_df.csv",encoding="utf-8", index=False, header=True)
```
