```python
wait_center = pd.read_csv('Wait_List_By_Center.csv')

wait_time = pd.read_csv('Wait_Times_By_Centers.csv')

donor_state = pd.read_csv('Transplants_by_Donor_Type state.csv')

cpra_score = pd.read_csv('CPRA.csv')
```

```python
wait_center = wait_center.drop(columns={'Unnamed: 2'})

wait_center = wait_center.rename(columns={'Unnamed: 0':'Center', 'Unnamed: 1':'Type'})

wait_center['Center'] = wait_center['Center'].fillna(method='ffill')

wait_center = wait_center.dropna(subset=['Type'])

wait_center
```

```python
wait_center = wait_center[['Center','Type','Kidney','Liver','Pancreas','Kidney / Pancreas','Heart']]

all_list = ['All Centers']

wait_center = wait_center[wait_center['Center'].isin(all_list)== False]

wait_center['Pancreas'] = wait_center['Pancreas'].astype(int)

wait_center
```

```python
wait_time = wait_time.drop(columns={'Unnamed: 2'})

wait_time = wait_time.drop(columns={'Unnamed: 3'})

wait_time = wait_time.rename(columns={'Unnamed: 0':'Center', 'Unnamed: 1':'State'})

wait_time['Center'] = wait_time['Center'].fillna(method='ffill')

wait_time = wait_time.dropna(subset=['State'])

wait_time
```

```python
wait_time = wait_time[['Center','State','Kidney','Liver','Pancreas','Kidney / Pancreas','Heart']]

time_all_list = ['All Centers']

wait_time = wait_time[wait_time['Center'].isin(all_list)== False]

wait_time['Pancreas'] = wait_time['Pancreas'].astype(int)

wait_time['Kidney'] = wait_time['Kidney'].str.replace(',','')

wait_time['Kidney'] = wait_time['Kidney'].astype(int)

type_list = ['All Center States']

wait_time['State'] = wait_time[wait_time['State'].isin(type_list) == False]

wait_time = wait_time.dropna(subset=['State'])

wait_time
```

```python
cpra_score = cpra_score.drop(columns={'Unnamed: 2'})

cpra_score = cpra_score.rename(columns={'Unnamed: 0':'Center', 'Unnamed: 1':'CPRA_Score'})

cpra_score['Center'] = cpra_score['Center'].fillna(method='ffill')

cpra_score = cpra_score.dropna(subset=['CPRA_Score'])

cpra_score
```

```python
cpra_score = cpra_score[['Center','CPRA_Score','Kidney','Liver','Pancreas','Kidney / Pancreas','Heart']]

cpra_all_list = ['All_CPRA']

cpra_score = cpra_score[cpra_score['Center'].isin(all_list)== False]

cpra_score['Pancreas'] = cpra_score['Pancreas'].astype(int)

cpra_score['Kidney'] = cpra_score['Kidney'].str.replace(',','')

cpra_score['Kidney'] = cpra_score['Kidney'].astype(int)

cpra_score
```

```python
import pandas as pd
import json

map_df = pd.read_json('mapjson.json')

map_df.head()

```

```python
import json
# load data using Python JSON module
with open('mapjson.json','r') as f:
    data = json.loads(f.read())
# Flatten data
map_df = pd.json_normalize(data, record_path =['locations'])


center_df = map_df.copy()

center_df = center_df.rename(columns={'All_Centers_Hospital_Name':'Center'})

center_df
```

```python
merge_time = center_df.merge(center_time, on='Center', how='inner')

merge_time
```

```python
final_df = cpra_score.merge(merge_time, on='Center',how='inner')

final_df = final_df.rename(columns={'Kidney_x':'Kidney_CPRA','Liver_x':'Liver_CPRA','Pancreas_x':'Pancreas_CPRA',
                                  'Kidney / Pancreas_x':'Kidney_Pancreas_CPRA'})

no_all_list = ['All Types']

Cpra_list = ['All CPRA']

final_df = final_df[final_df['Type'].isin(no_all_list) == False]

final_df = final_df[final_df['CPRA_Score'].isin(Cpra_list) == False]

final_df = final_df.dropna(subset=['Type'])

final_df
```
