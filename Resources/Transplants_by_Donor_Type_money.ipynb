```python
import pandas as pd
```

```python
csv_path_donor_type_money = "Transplants_by_Donor_Type_money.csv"

donortype_money_df = pd.read_csv(csv_path_donor_type_money, encoding="utf-8")
```

```python
donortype_money_df.head(50)
```

```python
#Fix names

donortype_money_df = donortype_money_df.rename(columns={"Unnamed: 0": "Donor_Type",
                                                        "Unnamed: 1": "Organ",
                                                        "Unnamed: 2": "Primary_Payer"
                                                       })

del donortype_money_df["Unnamed: 3"]
```

```python
donortype_money_df.head(50)
```

```python
#Remove unused organs

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "All Organs"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "NaN"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Lung"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Heart / Lung"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Intestine"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Abdominal Wall"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Head & Neck: Craniofacial"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Head & Neck: Scalp"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Head & Neck: Larynx"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "GU: Penile"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "GU: Uterus"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Upper Limb: Bilateral"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Upper Limb: Unilateral"]

donortype_money_df = donortype_money_df[donortype_money_df.Organ != "Bilateral Limb, Face"]

donortype_money_df.head(50)
```

```python
donortype_money_df.to_csv("Cleaned/Transplants_by_Donor_Type_money_clean.csv",encoding="utf-8", index=False, header=True)
```
