```python
import pandas as pd
```

```python
csv_path_wtbc = "Wait_Times_By_Centers.csv"

wtbc_df = pd.read_csv(csv_path_wtbc, encoding="utf-8")
```

```python
wtbc_df.head(50)
```

```python
wtbc_df = wtbc_df.fillna(method='ffill')
```

```python
wtbc_df = wtbc_df.rename(columns={"Unnamed: 0": "Center",
                                         "Unnamed: 1": "States",
                                         "Unnamed: 2": "Time_Waited"})
```

```python
#Split code and center name

wtbc_df[["Center_Code","Center_Name"]] = wtbc_df["Center"].str.split(n=1, expand=True)
```

```python
wtbc_df.tail(50)
```

```python
wtbc_df = wtbc_df[["Center","Center_Code","Center_Name", "States", "Time_Waited", "Kidney", "Liver", "Pancreas", "Kidney / Pancreas", "Heart"]]
```

```python
wtbc_df["Center_Code"] = wtbc_df["Center_Code"].replace(["All"],"All Centers")

wtbc_df["Center_Name"] = wtbc_df["Center_Name"].replace(["Centers"],"All Centers")
```

```python
wtbc_df.to_csv("Cleaned/wait_time_by_centers_df_clean.csv",encoding="utf-8", index=False, header=True)
```
