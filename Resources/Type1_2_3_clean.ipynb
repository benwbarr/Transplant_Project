```python
import pandas as pd
```

-------------------------------------------------------------------------------------------------------------------------------

1: All Donor Types

-------------------------------------------------------------------------------------------------------------------------------

```python
csv_path_type1 = "Transplants_by_Donor_Type_1_All_Donor_Types.csv"

type1_df = pd.read_csv(csv_path_type1, encoding="utf-8")

type1_df.head(30)
```

```python
#Fix names

type1_df = type1_df.rename(columns={"Unnamed: 0": "Organ"})

del type1_df['Unnamed: 1']
```

```python
#Remove unused organs

type1_df = type1_df[type1_df.Organ != "All Organs"]

type1_df = type1_df[type1_df.Organ != "Lung"]

type1_df = type1_df[type1_df.Organ != "Heart / Lung	"]

type1_df = type1_df[type1_df.Organ != "Intestine"]

type1_df = type1_df[type1_df.Organ != "Abdominal Wall"]

type1_df = type1_df[type1_df.Organ != "Head & Neck: Craniofacial"]

type1_df = type1_df[type1_df.Organ != "Head & Neck: Scalp"]

type1_df = type1_df[type1_df.Organ != "Head & Neck: Larynx"]

type1_df = type1_df[type1_df.Organ != "GU: Penile"]

type1_df = type1_df[type1_df.Organ != "GU: Uterus"]

type1_df = type1_df[type1_df.Organ != "Lung"]

type1_df = type1_df[type1_df.Organ != "Upper Limb: Bilateral"]

type1_df = type1_df[type1_df.Organ != "Upper Limb: Unilateral"]

type1_df = type1_df[type1_df.Organ != "Bilateral Limb, Face"]

type1_df.head(50)
```

```python
type1_df.to_csv("Cleaned/Transplants_by_Donor_Type_1_All_Donor_Types_df_clean.csv",encoding="utf-8", index=False, header=True)
```
