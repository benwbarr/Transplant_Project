```python
import pandas as pd
```

```python
csv_path_travel = "Travel_filled.csv"

travel_df = pd.read_csv(csv_path_travel, encoding="utf-8")
```

```python
#Clean Columns

travel_df = travel_df[["Unnamed: 0", "Unnamed: 1", "Kidney", "Liver", "Pancreas", "Kidney / Pancreas", "Heart"]]
```

```python
travel_df = travel_df.rename(columns={"Unnamed: 0": "Center",
                                     "Unnamed: 1": "State"})
```

```python
travel_df.head(30)
```

```python
#Split code and center name

travel_df[['Center_Code','Center_Name']] = travel_df["Center"].str.split(n=1, expand=True)
```

```python
travel_df.tail(50)
```

```python
travel_df = travel_df[["Center", "Center_Code", "Center_Name", "Kidney", "Liver", "Pancreas", "Kidney / Pancreas", "Heart"]]
```

```python
travel_df["Center_Code"] = travel_df["Center_Code"].replace(["All"],"All Centers")

travel_df["Center_Name"] = travel_df["Center_Name"].replace(["Centers"],"All Centers")
```

```python
travel_df.to_csv("Cleaned/Travel_df_clean.csv",encoding="utf-8", index=False, header=True)
```
